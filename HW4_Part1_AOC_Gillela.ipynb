```python
data_file = 'day1_hw4/data/input.txt'
with open(data_file, 'r') as input_file:
     numbers = input_file.read().splitlines()
```

```python
numbers
```

```python
numbers = [int(number) for number in numbers]
```

```python
numbers
```

```python
increase_count = 0  # to count how many times it increases

for i in range(1, len(numbers)):
    if numbers[i] > numbers[i - 1]:
        increase_count = increase_count + 1
    
print(f"\nTotal increases: {increase_count}")
```

```python
list_of_numbers = []

for i in range(len(numbers) - 2):  # stops at len - 3 so we donâ€™t go out of bounds
    list_of_number = numbers[i] + numbers[i+1] + numbers[i+2]
    list_of_numbers.append(list_of_number)

#print(list_of_numbers)

increase_count1 = 0  # to count how many times it increases

for i in range(1, len(list_of_numbers)):
    if list_of_numbers[i] > list_of_numbers[i - 1]:
        increase_count1 += 1
    
print(f"\nTotal increases: {increase_count1}")
```
